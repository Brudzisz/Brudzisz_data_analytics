```python
import numpy as np
from cmdstanpy import CmdStanModel
import pandas as pd
import matplotlib.pyplot as plt

#Observed data
x_obs = [-10, -8, -6, -4, -2, 0, 2, 4, 6, 8, 10]
y_obs = [0.328572824089476, 4.20607004111644, 1.35507551134795,
0.161608755204364, -5.42320349780782, -3.05851276224202, -0.0764172642034502,
-4.55218472276499, -0.902226297922731, -5.8609833528976, -1.05854090910473]

x_predict = np.linspace(-11,11,551,endpoint=True)
observed_idx = [26, 76, 126, 176, 226, 276, 326, 376, 426, 476, 526]

alpha = 3
rho = 5.5
sigma = 2

model = CmdStanModel(stan_file='gaussian_process.stan')

data1 = {
    'N_predict': len(x_predict),
    'x_predict': x_predict,
    'N_obs': len(observed_idx),
    'y_obs': y_obs,
    'observed_idx': observed_idx,
    'rho': rho,
    'alpha': alpha,
    'sigma': sigma
}

data2 = {
    'N_predict': len(x_predict),
    'x_predict': x_predict,
    'N_obs': len(observed_idx),
    'y_obs': y_obs,
    'observed_idx': observed_idx,
    'rho': 1,
    'alpha': 1,
    'sigma': 1
}

data3 = {
    'N_predict': len(x_predict),
    'x_predict': x_predict,
    'N_obs': len(observed_idx),
    'y_obs': y_obs,
    'observed_idx': observed_idx,
    'rho': 10,
    'alpha': 5,
    'sigma': 1
}

datas = [data1, data2, data3]
```

```python
result1 = model.sample(data=datas[0], chains = 4)
result2 = model.sample(data=datas[1], chains = 4)
result3 = model.sample(data=datas[2], chains = 4)
```

```python
#1. plot few (about 5-10) samples generated from model
#2. plot mean value from model with its confidence interval (standar deviation, using errorbar function)
#3. Also on each plot place points with obsserved data. Repeat the steps for two other parameters sets of your choice
# df = result.draws_pd()
# print(df)



prediction1 = result1.stan_variable('y_predict')
prediction2 = result2.stan_variable('y_predict')
prediction3 = result3.stan_variable('y_predict')
predictions = [prediction1, prediction2, prediction3]
samples = [1, 10, 100, 2, 20, 200]


for j in range(len(predictions)):
    n = 0
    title = "Rho="+str(datas[j]['rho'])+"; alpha="+str(datas[j]['alpha'])+"; sigma="+str(datas[j]['sigma'])

    for i in samples:
        n += 1
        sample = predictions[j][:, i]
        mean = np.mean(sample)
        std = np.std(sample)
        plt.plot(sample)
        plt.scatter(observed_idx, y_obs, color='k', zorder = 3)
        plt.errorbar(range(len(sample)), [mean]*len(sample), std)
        plt.errorbar(range(len(sample)), [mean]*len(sample), color = 'r')
        plt.title(title + '; Sample ' + str(n))
        plt.legend(["Sample", "Observed data", "Confidence interval", "Mean value"], loc = "upper right")
        plt.show()
```
